### For differential equations the form is ay''+by'+cy=dt
#### The intial for x1 = a_1 , x2 = a_2

```python
def number_of_steps(to, tf, h):
    n = (tf-to)/h
    return n
```

```python
def steps_time(to , tf, h):
    t = []
    n = []
    for i in range(int(number_of_steps(to, tf, h))+1):
        t.append(i*h+to)
        n.append(int(i))
    return t , n
```

```python
def x_1_value(h, t,  x_1, x_2):
    return x_1 + h*x_2
```

### You  can adjust the formula for X_2 to modify to fit the formula of expects, here is a dt formula but you can easily change it

```python
def x_2_value(h, t, x_1, x_2, a,b,c,d):
    return x_2 + h*((d*t-b*x_2-c*x_1)/a)
```

### Method input is time intial, time final, step size, y(t) intial, y'(t) intial, 

```python
def eulers_method(to, tf, h ,a_1, a_2, a,b,c,d):
    x1list = []
    x2list = []
    t = steps_time(to , tf, h)[0]
    n = steps_time(to, tf, h)[1]
    x1list.append(a_1)
    x2list.append(a_2)

    for i in range(len(n)):
        x1 = x_1_value(h, t[i], x1list[i], x2list[i])
        x1list.append(x1)
        
        x2 = x_2_value(h, t[i], x1list[i], x2list[i], a,b,c,d)
        x2list.append(x2)
        
    return n, t ,x1list, x2list
```

```python
df = pd.DataFrame(eulers_method(0,12,0.5,2,-1,1,-3,2, 1),['Step','Time','x1','x2'])
```
